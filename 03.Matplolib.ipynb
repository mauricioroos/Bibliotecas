```python
# histograma analisar a destribuicao de notas de uma turma de 500 alunos
```

```python
import numpy as np
np.random.seed(42)
notas = np.random.normal(70, 10, 500)
```

```python
plt.figure(figsize=(8,5))
plt.hist(notas, bins=50, edgecolor='black')
plt.title("Distribuição de notas da turma")
plt.xlabel('Notas')
plt.xlabel('Número de alunos')
plt.axvline(notas.mean(), color='red',linestyle='--', linewidth='2')
plt.legend(['Média das notas'])

plt.grid(True, linestyle='--', alpha=0.5)
plt.ylim(0,35)
plt.show()
```

```python
## grafico pizza para anilisar a composição de uma carteira de investimentos
```

```python
ativos = ['Açõe', 'Renda Fixa', 'Ouro', 'Criptomoeda']
valores = [45, 30, 15, 10]
```

```python
plt.figure(figsize=(5,5))
explode =(0.1, 0, 0, 0)

plt.pie(valores, labels=ativos, startangle=90,autopct='%1.1f%%', explode=explode)
plt.title('composição da carteira de investimentos') 
plt.axis('equal')
plt.show()
```

```python
## grafico de disperção para analisar a relação da altura e peso das pessoas
```

```python
np.random.seed(0)
alturas = np.random.normal(170, 10, 50)
peso = (alturas * 0.5) + np.random.normal (0, 5, 50)
```

```python
plt.figure(figsize=(8,4))
plt.scatter(alturas, peso, color='red', s=20, alpha=0.5)
plt.title("Relação entre altura e peso")
plt.xlabel("Altura (cm)")
plt.ylabel("Peso (kg)")
plt.grid(True, linestyle='--', alpha=0.5)

plt.show()



```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('titanic.csv')
```

```python
sobreviventes = df['Survived'].value_counts()
sobreviventes
```

```python
sobreviventes_por_sexo = df.groupby('Sex')['Survived'].sum()
sobreviventes_por_sexo
```

```python
contagem_sobreviventes_sexo = df.groupby(['Sex', 'Survived']).size().unstack()
```

```python
contagem_sobreviventes_sexo.plot(kind='bar', figsize=(10, 6), rot=0)

plt.title('Contagem de Sobreviventes e Não Sobreviventes por Gênero')
plt.xlabel('Gênero')
plt.ylabel('Número de Sobreviventes')
plt.xticks(ticks=[0, 1], labels=['Feminino', 'Masculino'])
plt.legend(title='Sobrevivência', labels=['Não Sobreviventes', 'Sobreviventes'])

plt.show()
```
